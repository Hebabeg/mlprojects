```python
import seaborn as sns #For Graphs
import matplotlib.pyplot as plt #For Plots and Graphs
import numpy as np #For Operations
import pandas as pd #For Data Frames
from sklearn.model_selection import train_test_split #No need to manual split now, this model will split the data
from sklearn.ensemble import RandomForestRegressor #Method
from sklearn import metrics #Model Evaluation (error score and accuracy score)


```

```python
gold = pd.read_csv('/content/gld_price_data.csv')
```

```python
plt.figure(figsize= (8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot=True, annot_kws={'size':8}, cmap='Reds')
```

```python
print(correlation['GLD'])
```

```python
sns.distplot(gold['GLD'],color='blue')
```

```python
X = gold.drop(['Date','GLD'],axis=1)
Y = gold['GLD']
print(Y)
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=2)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

```python
test_prediction = regressor.predict(X_test)
```

```python
print(test_prediction)
```

```python
Y_test = list(Y_test)
plt.plot(Y_test, color='red', label = 'Actual Value')
plt.plot(test_prediction, color='yellow', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```

```python
R_squared_error = metrics.r2_score(Y_test, test_prediction)
print("R squared error : ", R_squared_error)
```
